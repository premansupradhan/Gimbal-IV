```python
import matplotlib.pyplot as plt
import math
import numpy as np

ang = list(np.arange(0, (2 * math.pi), 0.02))
print(ang)

fig = plt.figure()
ax = plt.axes(projection='polar')

for i in ang:
    a = 10
    b = i
    c = [0, b]
    d = [0, a]
    line = ax.plot(c, d, linestyle='-')  # Store the plot object
    plt.pause(0.05)
    line[0].remove()  # Clear the previous line drawn

plt.show()
```
